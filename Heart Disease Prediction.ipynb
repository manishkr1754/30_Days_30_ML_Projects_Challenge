## 1) Understanding Problem Statement
---

Heart disease is a critical health issue affecting a significant portion of the population worldwide. Early detection and accurate prediction of heart disease can be life-saving and can improve the overall quality of healthcare. Machine learning models can play a crucial role in predicting the likelihood of heart disease based on various medical and lifestyle factors. 

The project falls under **Classification Machine Learning Problem**. The goal of this project is **to develop a heart disease prediction model to predict the likelihood of an individual having heart disease based on their medical history, lifestyle factors and diagnostic test results**.

## 2) Understanding Data
---

The project uses **Heart Disease Data** which contains several variables (independent variables) and the outcome variable or dependent variable.
The varibales in the dataset are as follows:

- age: Age
- sex: (1 = male; 0 = female)
- cp: Chest pain (typical=0, asymptotic=1, nonanginal=2, nontypical=3)
- trestbps: Resting blood pressure 
- chol: Serum cholestoral in mg/dl 
- fbs: Fasting blood sugar > 120 mg/dl (1 = true; 0 = false)
- restecg: Resting electrocardiographic results
- exang: Exercise induced angina (1 = yes; 0 = no)
- oldpeak: ST depression induced by exercise relative to rest
- slope: Slope of the peak exercise ST segment 
- ca: Number of major vessels colored by flourosopy (0 - 3)
- thal: (3 = normal; 6 = fixed defect; 7 = reversable defect)
- target: AHD - Diagnosis of heart disease (1 = yes; 0 = no)

## 5) Model Building
---

### Creating Feature Matrix (Independent Variables) & Target Variable (Dependent Variable)

```python
# separating the data and labels
X = heart_disease_data.drop(columns = ['target'], axis=1) # Feature matrix
y = heart_disease_data['target'] # Target variable
```

```python
X
```

```python
y
```

### Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
standardized_data
```

```python
X = standardized_data
```

```python
X
```

### Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=45)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(y.shape, y_train.shape, y_test.shape)
```

### Model Comparison : Training & Evaluation

```python
models = [LogisticRegression, SVC, DecisionTreeClassifier, RandomForestClassifier]
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []

for model in models:
    classifier = model().fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    
    accuracy_scores.append(accuracy_score(y_test, y_pred))
    precision_scores.append(precision_score(y_test, y_pred))
    recall_scores.append(recall_score(y_test, y_pred))
    f1_scores.append(f1_score(y_test, y_pred))
```

```python
classification_metrics_df = pd.DataFrame({
    "Model": ["Logistic Regression", "SVM", "Decision Tree", "Random Forest"],
    "Accuracy": accuracy_scores,
    "Precision": precision_scores,
    "Recall": recall_scores,
    "F1 Score": f1_scores
})

classification_metrics_df.set_index('Model', inplace=True)
classification_metrics_df
```

### Inference

In the context of Heart Disease Prediction, 
- The **SVM (Support Vector Machine)** model outperforms other models in predicting heart disease, achieving an accuracy of 83.61% with a well-balanced precision of 78.05% and high recall of 96.97%. This indicates its capability to accurately identify individuals at risk of heart disease while minimizing false alarms. The model's F1 score of 86.49% demonstrates its strong overall performance, making it a promising choice for heart disease prediction in healthcare applications.
