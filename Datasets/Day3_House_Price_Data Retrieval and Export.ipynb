<center><h1> House Price Data Retrieval and Export </h1></center>

The project uses California Housing Data available in sklearn library. 

- **Data Retrieval:** We will fetch California Housing Price Data using `sklearn.datasets.fetch_california_housing()`. This data is available within scikit-learn and we can load it into your Python environment using the provided function.So, we can work with the dataset directly without the need to export it to a .csv file.

- **Exporting as .csv:** We will export this data as .csv for future use case and sharing using `dataframe_name.to_csv('file_name.csv', index=False)`.

```python
# Importing necessary libraries
import pandas as pd
import numpy as np

import sklearn.datasets
```

```python
house_price_data = sklearn.datasets.fetch_california_housing()
```

```python
house_price_df = pd.DataFrame(house_price_data.data, columns = house_price_data.feature_names)
```

```python
house_price_df.head()
```

```python
house_price_df['Price'] = house_price_data.target
```

```python
house_price_df.head()
```

```python
# Exporting as .csv
house_price_df.to_csv('Day3_House_Price_data.csv', index=False)
```
